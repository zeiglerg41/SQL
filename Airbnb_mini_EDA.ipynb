#### Building on the previous question where you identified the top 3 Neighbourhoods within each of the three neighbourhood_groups based off Revenues, filter the Airbnb Dataframe to include only these neighbourhoods. 
    
Upon completion of this, identify the  <b> top average revenue generating room type</b> for each of the nine neighbourhoods and plot this out in a Bar Chart.
    
    

```python
# Identify the top 3 revenue-generating neighbourhoods for each neighbourhood group
top_3_neigh = df[df['neighbourhood_group'].isin(['Manhattan', 'Brooklyn', 'Queens'])]
grouped = top_3_neigh.groupby(['neighbourhood_group', 'neighbourhood']).agg({'revenue': 'sum'}).reset_index()
top_neigh = grouped.groupby('neighbourhood_group').apply(lambda x: x.nlargest(3, 'revenue')).reset_index(drop=True)

# Filter the original dataframe to include only the top neighbourhoods
filtered_df = df[df['neighbourhood'].isin(top_neigh['neighbourhood'])]

# Step 3: Calculate the average revenue for each room type and identify the top one for each neighbourhood
room_avg_revenue = filtered_df.groupby(['neighbourhood_group', 'neighbourhood', 'room_type']).agg({'revenue' : 'mean'}).reset_index()
top_room_by_neigh = room_avg_revenue.groupby(['neighbourhood_group', 'neighbourhood']).apply(lambda x: x.nlargest(1, 'revenue')).reset_index(drop=True)
```

```python
# Create a combined column for Neighbourhood and Room Type
top_room_by_neigh['neigh_room_type'] = (top_room_by_neigh['neighbourhood'] + ' (' + top_room_by_neigh['room_type'] + ')')

# Sort revenues descending
sorted_top_room = top_room_by_neigh.sort_values('revenue', ascending = False)

# Plot the bar chart using the combined column
ax3 = sorted_top_room.plot(kind='bar', x='neigh_room_type', y='revenue', figsize=(8,6))
plt.title('Popular Airbnb Neighbourhoods by Room Type')
plt.xlabel('Neighbourhood/Room Type')
plt.ylabel('Revenue')

# Add data labels
for p in ax3.patches:
    height = p.get_height()
    ax3.text(p.get_x() + p.get_width() / 2, height, f'{height/1e3:.2f}K', 
            ha="center", va='bottom')
plt.tight_layout()
plt.show()
```
